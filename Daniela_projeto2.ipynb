### Tabela de Comando SQL

| **Comando SQL**        | **Descrição**                                                                 | **Exemplo de Uso**                                                                                     |
|------------------------|-------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------------------|

| `HAVING`                | Filtra grupos após o uso de `GROUP BY`.                                       | ```SELECT curso, COUNT(*) FROM Alunos GROUP BY curso HAVING COUNT(*) > 5;```                             |

### Cenário: Análise da Jornada e Experiência do Cliente

#### Descrição do Cenário:


Uma rede de hotéis está interessada em entender melhor a jornada dos seus clientes no site, bem como a experiência de reseva e estadia. A empresa tem informações sobre:

  - **Navegação:** O cliente acessa o site e visualiza diferentes páginas.

  - **Reservas:** O cliente faz uma reserva de quarto em um hotel.

  - **Avaliações:** O cliente avalia sua experiência após a estadia.

A rede quer analisar como diferentes perfis de clientes avaliam a experiência de estadia, comparando esses resultados com todo ofunil até chegar a estadia, desde a navegação até a avaliação do hotel.

#### Objetivo:

  - Entender a jornada do cliente desde a navegação no site até a reserva e sua experiência final.

  - Perguntas para análise:
  1. Qual o tempo médio que os clientes gastam no site antes de fazer uma reserva?

  2. Qual o percentual de clientes que realizam reservas após a navegação?

3. Existe alguma correlação entre o tempo gasto no site e a avaliação dada?

````python
import sqlite3 as sql
````

````python
# Conectar ao banco de dados SQLite
conexao = sql.connect('jornada_cliente.db')
cursor = conexao.cursor()
````

````python
# Criar tabelas Reserva, Navegacao, Clientes e Avaliacao
# Cliente
cursor.execute('''
  CREATE TABLE IF NOT EXISTS Clientes (
    id_cliente INTEGER PRIMARY KEY,
    nome TEXT,
    email TEXT,
    idade INTEGER,
    genero TEXT
  )
  ''')
````

````python
# Criar tabelas Reserva, Navegacao, Clientes e Avaliacao
# Reserva
cursor.execute('''
  CREATE TABLE IF NOT EXISTS Reserva (
    id_reserva INTEGER PRIMARY KEY,
    id_cliente INTEGER,
    data_check_in DATE,
    data_check_out DATE,
    valor_total REAL,
    FOREIGN KEY (id_cliente) REFERENCES Clientes(id_cliente)
  )
  ''')
````

````python
# Criar tabelas Reserva, Navegacao, Clientes e Avaliacao
# Navegacao
cursor.execute('''
  CREATE TABLE IF NOT EXISTS Navegacao (
    id_navegacao INTEGER PRIMARY KEY,
    id_cliente INTEGER,
    pagina_visitada TEXT,
    tempo_gasto INTEGER,
    data_acesso DATE,
    FOREIGN KEY (id_cliente) REFERENCES Clientes(id_cliente)
  )
  ''')
````

````python
# Criar tabelas Reserva, Navegacao, Clientes e Avaliacao
# Avaliacao
cursor.execute('''
  CREATE TABLE IF NOT EXISTS Avaliacao (
    id_avaliacao INTEGER PRIMARY KEY,
    id_cliente INTEGER,
    nota INTEGER,
    comentario TEXT,
    FOREIGN KEY (id_cliente) REFERENCES Clientes(id_cliente)
  )
  ''')
````

````python
# Função para ler dados CSV e inseri-los nas tabelas SQLite
import csv

def inserir_dados_csv(arquivo, tabela, colunas, num_colunas):
    with open(arquivo, newline='') as file:
        reader = csv.reader(file)
        next(reader)  # Pular o cabeçalho
        dados = [tuple(row) for row in reader]
        placeholder = ', '.join(['?'] * num_colunas)  # Gerar placeholders de acordo com o número de colunas
        query = f"INSERT OR IGNORE INTO {tabela} ({colunas}) VALUES ({placeholder})"
        cursor.executemany(query, dados)
````

````python
inserir_dados_csv('clientes.csv', 'Clientes', 'id_cliente, nome, email, idade, genero', 5)
inserir_dados_csv('navegacao.csv', 'Navegacao', 'id_navegacao, id_cliente, pagina_visitada, tempo_gasto, data_acesso', 5)
inserir_dados_csv('reserva.csv', 'Reserva', 'id_reserva, id_cliente, data_check_in, data_check_out, valor_total', 5)
inserir_dados_csv('avaliacoes.csv', 'Avaliacao', 'id_avaliacao, id_cliente, nota, comentario', 4)
````

````python
# Salvar alterações e fechar conexão
conexao.commit()
#conexao.close()
print("Dados inseridos com sucesso no banco de dados!")
````

# Consultas
Qual o tempo médio que os clientes gastam no site antes de fazer uma reserva?

Qual o percentual de clientes que realizam reservas após a navegação?

Existe alguma correlação entre o tempo gasto no site e a avaliação dada?

````python
# Consultar Qual o tempo médio que os clientes gastam no site antes de fazer uma reserva?
cursor.execute('''
  SELECT AVG(tempo_gasto)
  FROM Navegacao
  WHERE id_cliente IN (
    SELECT id_cliente
    FROM Reserva
  )
''')
````

````python
# Imprimir o resultado
resultado = cursor.fetchall()[0][0]
print(f"O tempo médio que os clientes gastam no site antes de fazer uma reserva é de {resultado:.2f} segundos.")
````

````python
# Criar a tabela para armazenar o tempo médio que os clientes gastam no site antes de fazer uma reserva
cursor.execute('''
  CREATE TABLE IF NOT EXISTS TempoMedioReserva (
    tempo_medio REAL,
    FOREIGN KEY (tempo_medio) REFERENCES TempoMedio(tempo_medio)
  )
''')
````

````python
# Consultar Qual o percentual de clientes que realizam reservas após a navegação?
cursor.execute('''
  SELECT (COUNT(DISTINCT id_cliente) * 100.0) / (SELECT COUNT(DISTINCT id_cliente) FROM Navegacao) AS percentual
  FROM Reserva
  WHERE id_cliente IN (
    SELECT id_cliente
    FROM Navegacao
  )
''')
````

````python
# Imprimir o resultado
resultado = cursor.fetchall()[0][0]
print(f"O percentual de clientes que realizam reservas após a navegação é de {resultado:.2f}%.")
````

````python
# Criar a tabela para armazenar o percentual de clientes que fizeram reserva após a navegaçao
cursor.execute('''
  CREATE TABLE IF NOT EXISTS PercentualReserva (
    percentual REAL,
    FOREIGN KEY (percentual) REFERENCES Percentual(percentual)
  )
''')
````

````python
# Consultar Se Existe alguma correlação entre o tempo gasto no site e a avaliação
cursor.execute('''
  SELECT Navegacao.tempo_gasto, Avaliacao.nota
  FROM Navegacao
  JOIN Avaliacao ON Navegacao.id_cliente = Avaliacao.id_cliente
''')
````

````python
# Imprimir o resultado apenas 5 linhas
resultado = cursor.fetchall()
for linha in resultado:
    print(f"Tempo Gasto: {linha[0]} segundos - Nota: {linha[1]}")
````

````python
# Criar a tabela para armazenar a correlação entre tempo gasto e avaliacao
cursor.execute('''
  CREATE TABLE IF NOT EXISTS Correlacao (
    tempo_gasto INTEGER,
    nota INTEGER,
    FOREIGN KEY (tempo_gasto) REFERENCES Navegacao(tempo_gasto)
    FOREIGN KEY (nota) REFERENCES Avaliacao(nota)
  )
''')
````

````python
# Salvar as alterações e fechar a conexão
conexao.commit()
conexao.close()
````
